```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
```

```python
### Demographic Analysis 

r = requests.get("https://www.worldometers.info/coronavirus/coronavirus-age-sex-demographics/")
html = r.text
dfs = pd.read_html(html)
```

```python
# Age

df_age = dfs[0]
df_age.columns = df_age.iloc[0]
df_age = df_age[1:].set_index("AGE")
df_age
```

```python
# Gender

df_gender = dfs[1]
df_gender.columns = df_gender.iloc[0]
df_gender = df_gender[1:].set_index("SEX")
df_gender
```

```python
# Pre-existing Medical Conditions

df_health = dfs[2]
df_health.columns = df_health.iloc[0]
df_health = df_health[1:].set_index("PRE-EXISTING CONDITION")
df_health
```

```python
### Covid-19 Death Toll 

r = requests.get("https://www.worldometers.info/coronavirus/")
html_text = r.text
dfs = pd.read_html(html_text)

df = dfs[0]

df = df[["Country,Other", "TotalCases", "TotalDeaths", "TotalRecovered"]]
df = df.rename(columns={"Country,Other":"Country"})
df = df.set_index("Country")
df = df.fillna(0)
df["DeathToll"] = df["TotalDeaths"] / df["TotalCases"]

df
```
